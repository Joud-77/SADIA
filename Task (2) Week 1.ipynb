### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
sns.set(style="whitegrid")

# Create a figure and axes
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20, 15))

# Flatten the axes array for easier iteration
axes = axes.flatten()

# List of columns for which to create histograms
columns = ['Datetime','TrafficIndexLive']

# Plot histogram for each column
for i, col in enumerate(columns):
    df2[col].hist(ax=axes[i], bins=20)
    axes[i].set_title(f'Histogram of {col}')
    axes[i].set_xlabel('Datetime')
    axes[i].set_ylabel('TrafficIndexLive')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.suptitle('distribution of key variables', y=1.05)  # Adjust title position
plt.show()
```

```python
#anuthor plot 
plt.title('distribution of key variables')
plt.xlabel('TrafficIndexLive')
plt.ylabel('Datetime')
plt.xticks(df2['TrafficIndexLive'])  
plt.legend()  # Show legend with labels
plt.grid(True)  # Show grid for better readability
plt.tight_layout()
```

# Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
# Scatter plots to explore relationships between 'Age_band_of_driver' and ' Accident_severity'
plt.figure(figsize=(15, 10))
sns.scatterplot(data=df2, x='TrafficIndexLive', y='Weather_conditions', palette='tab10', edgecolor=None)
plt.title('TrafficIndexLive vs. Weather_conditions')
plt.show()
```
