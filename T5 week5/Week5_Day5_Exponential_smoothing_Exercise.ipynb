#**Double Exponential Smoothing**




```python
model_double = sm.tsa.Holt(data['Vehicles'])
model_double_fit = model_double.fit()
```

## Making predictions



```python
forecast_double = model_double_fit.forecast(6)
print(forecast_double)
```

#**Holt-Winters Seasonal Smoothing**






```python
model_triple = ExponentialSmoothing(data['Vehicles'],
                                    seasonal_periods=12,
                                    trend='add',
                                    seasonal='add') # Remove the extra 'data' argument

# Fit the model to the data
model_triple_fit = model_triple.fit()
```

## Making predictions



```python
forecast_triple = model_triple_fit.forecast(6)
print(forecast_triple)
```
